# Land use change model of Mozambique

## Run the model

```python
import sys, os
os.getcwd()
```

```python
from LU_Moz import LandUseChangeModel
from pcraster.framework import DynamicFramework, MonteCarloFramework
import Parameters
```

```python
nrOfTimeSteps = Parameters.getNrTimesteps()
nrOfSamples = Parameters.getNrSamples()
myModel = LandUseChangeModel()
dynamicModel = DynamicFramework(myModel,nrOfTimeSteps)
mcModel = MonteCarloFramework(dynamicModel, nrOfSamples)
```

```python
mcModel.run()
```
